```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('canada_per_capita_income.csv')
```

```python
X = dataset[['year']]
y = dataset['per capita income (US$)']
```

```python
model = LinearRegression()
# At this stage, the model has not learned anything.
```

```python
model.fit(X, y)
#  Now the model finds the best-fit line.
# Internally it computes Income = m × Year + c
 
```

```python
prediction = model.predict([[2020]])

print("Predicted Per Capita Income:", prediction[0])
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, model.predict(X), color='blue')
plt.title("Canada Per Capita Income")
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.show()
```

```python
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)

# Then the learned equation is Income=828.46×Year−1632210.76
```
