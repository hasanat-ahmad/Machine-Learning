```python
# Use wine dataset from sklearn.datasets to classify wines into 3 categories. Load the dataset and split it into test and train. After that train the model using Gaussian and Multinomial classifier and evaluate which performs better. Use the trained model to perform some predictions on test data.

# 1. Import required libraries
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn import metrics

# 2. Load the dataset
wine = load_wine()
X = wine.data       # 13 chemical features (alcohol, malic acid, ash, etc.)
y = wine.target     # 3 classes (0, 1, 2 -> three wine cultivars)

print("Feature names:", wine.feature_names)
print("Target names:", wine.target_names)
print("Shape of X:", X.shape)

# 3. Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 4. Train Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)

# 5. Train Multinomial Naive Bayes
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
y_pred_mnb = mnb.predict(X_test)

# 6. Evaluate both models
gnb_accuracy = metrics.accuracy_score(y_test, y_pred_gnb) * 100
mnb_accuracy = metrics.accuracy_score(y_test, y_pred_mnb) * 100

print("\nGaussian Naive Bayes Accuracy: {:.2f}%".format(gnb_accuracy))
print("Multinomial Naive Bayes Accuracy: {:.2f}%".format(mnb_accuracy))

# 7. Compare performance
if gnb_accuracy > mnb_accuracy:
    print("\nGaussian Naive Bayes performs better on this dataset.")
elif mnb_accuracy > gnb_accuracy:
    print("\nMultinomial Naive Bayes performs better on this dataset.")
else:
    print("\nBoth models perform equally well.")

# 8. Detailed evaluation (confusion matrix + classification report)
print("\nGaussianNB Confusion Matrix:\n", metrics.confusion_matrix(y_test, y_pred_gnb))
print("\nGaussianNB Classification Report:\n", metrics.classification_report(y_test, y_pred_gnb))

print("\nMultinomialNB Confusion Matrix:\n", metrics.confusion_matrix(y_test, y_pred_mnb))
print("\nMultinomialNB Classification Report:\n", metrics.classification_report(y_test, y_pred_mnb))

# 9. Make predictions on a few individual test samples
print("\nSample Predictions (Gaussian NB):")
for i in range(5):
    print("Actual:", wine.target_names[y_test[i]],
          " | Predicted:", wine.target_names[y_pred_gnb[i]])

print("\nSample Predictions (Multinomial NB):")
for i in range(5):
    print("Actual:", wine.target_names[y_test[i]],
          " | Predicted:", wine.target_names[y_pred_mnb[i]])
```
